```python
import pandas as pd  

data = {'Name': ['Alice', 'Bob', 'Charlie'], 
        'Age': [25, 30,35],  
        'Gender': ['F', 'M', 'M']}  

df = pd.DataFrame(data)

average_age = df['Age'].mean().item()

print(f"Average Age: {average_age}")
```

```python
import numpy as np
from scipy.stats import zscore

data = [1, 2, 3, 100, 5, 6, 7, 8, 9]
median = np.median(data).item()
z_scores = zscore(data)
data = [median if abs(z) > 2 else val for val, z in zip(data, z_scores)]
data
```
